```python
fig, axs = plt.subplots(ncols=2, figsize=(12, 5))
plt.rcParams["font.size"] = 15
axs[0].set_ylim([-0.4,8.4])
# axs[1].set_ylim([-0.3,8])
pg.plot_paired(data=df, dv='# of backward waves (30s)', within='Light', subject='index', ax=axs[0], boxplot=False, colors=['green', 'green', 'green'], pointplot_kwargs={'scale':0.8})
pg.plot_paired(data=df, dv='# of forward waves (30s)', within='Light', subject='index', ax=axs[1], boxplot=False, colors=['green', 'green', 'green'], pointplot_kwargs={'scale':0.8})
fig.suptitle('20~25 µW/$\mathsf{mm^2}$ ($n=2$)', fontsize=18)
plt.savefig("1.5mW.svg")
```
